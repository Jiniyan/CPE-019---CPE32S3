### 3.1 **Create a base model**


```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical

# Convert string labels to integers
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# Convert integer labels to one-hot encoded format
y_onehot = to_categorical(y_encoded)

# Now, y_onehot can be used as your target variable for training the model
```

```python
# define baseline model

# Convert string labels to integers
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# Convert integer labels to one-hot encoded format
y_onehot = to_categorical(y_encoded)

# Define a base model function
def baseline_model():
    # Initialize the sequential model
    model = Sequential()
    model.add(Dense(8, input_dim=X.shape[1], activation='relu'))
    model.add(Dense(2, activation='softmax'))

    # Compile the model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model

# Create the base model
model = baseline_model()

# Print the summary of the model
model.summary()

# Train the model using one-hot encoded labels
model.fit(X, y_onehot, epochs=10, batch_size=32)
```

```python
estimator = KerasClassifier(build_fn=baseline_model, epochs=200, batch_size=5, verbose=0)
```

###  3.2 **Evaluate the model with k-fold cross validation**


```python
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
```

```python
results = cross_val_score(estimator, X, dummy_y, cv=kfold)
print("Accuracy: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

### 3.3 **Improve the accuracy of your model by applying additional hidden layers**


```python
# Convert string labels to integers
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# Convert integer labels to one-hot encoded format
y_onehot = to_categorical(y_encoded)


def improved_model():
    model = Sequential()
    model.add(Dense(8, input_dim=X.shape[1], activation='relu'))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(2, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model

# Create the improved model
model = improved_model()

# Print the summary of the model
model.summary()

# Train the model using one-hot encoded labels
model.fit(X, y_onehot, epochs=10, batch_size=32)
```

```python
estimator2 = KerasClassifier(build_fn=improved_model, epochs=100, batch_size=5, verbose=0)
results2= cross_val_score(estimator2, X, dummy_y, cv=kfold)
print("Improved Model Accuracy: %.2f%% (%.2f%%)" % (results2.mean()*100, results2.std()*100))
```

```python
# define base mode
def baseline_model2():
# create model
  model2 = Sequential()
  model2.add(Dense(10, input_dim=X.shape[1], kernel_initializer='normal', activation='relu'))
  model2.add(Dense(1, kernel_initializer='normal'))
  model2.compile(loss='mean_squared_error', optimizer='adam')

  return model2
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
# Create pipeline
estimators = [
    ('impute', SimpleImputer()),      # Impute missing values
    ('standardize', StandardScaler()),# Standardize features
    ('mlp', KerasRegressor(build_fn=baseline_model2, epochs=100, batch_size=5, verbose=0))
]
pipeline = Pipeline(estimators)
```

```python
#evaluate model
kfold = KFold(n_splits=10)
```

```python
results = cross_val_score(pipeline, X, Y, cv=kfold, scoring='neg_mean_squared_error')
print("Baseline: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

### 4.3 Show tuning of layers and neurons (see evaluating small and larger networks)


```python
# Small Network
def small_model():
    # create model2
    model2 = Sequential()
    model2.add(Dense(10, input_shape=(13,), kernel_initializer='normal', activation='relu'))
    model2.add(Dense(1, kernel_initializer='normal'))

    # Compile model2
    model2.compile(loss='mean_squared_error', optimizer='adam')
    return model2

# evaluate model2
small_estimator = KerasRegressor(model2=small_model, epochs=100, batch_size=5, verbose=0)
```

```python
small_results = cross_val_score(small_estimator, X, Y, cv=kfold, scoring='neg_mean_squared_error')
print("Small Network: %.2f (%.2f) MSE" % (small_results.mean(), small_results.std()))
```

```python
# Larger Network
def large_model():
    # create model
    model = Sequential()
    model.add(Dense(20, input_shape=(13,), kernel_initializer='normal', activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))

    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# evaluate model
large_estimator = KerasRegressor(model=large_model, epochs=100, batch_size=5, verbose=0)
```

```python
large_results = cross_val_score(large_estimator, X, Y, cv=kfold, scoring='neg_mean_squared_error')
print("Large Network: %.2f (%.2f) MSE" % (large_results.mean(), large_results.std()))
```
